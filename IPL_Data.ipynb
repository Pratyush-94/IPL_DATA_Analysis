```python
# CSK Rexords at M. A. Chidambaram Stadium matches
Chennai = matchinfodata[matchinfodata['city'] == 'Chennai']
Chennai
```

```python
#Chennai WON - M. A. Chidambaram Stadium 
Chennaiwin = Chennai[Chennai['winner'] == 'Chennai Super Kings']
Chennaiwin

```

```python
# Toss Won and  Field First Decision


tosswon_field = Chennai[Chennai['toss_decision'] == 'field']
tosswon_field
```

```python
#CSK Summary after Toss WON and Field First and Result WIN
tosswon_field= Chennai[(Chennai['toss_decision'] == 'field') & (Chennai['winner'] == 'Chennai Super Kings')]
print(tosswon_field)
```

```python
# Total count of CSK  Toss WON and Field First and Result LOSS
tossfieldlostcount = len(tossfieldlost)
tossfieldlostcount
```

```python
# CSK Summary after Toss WON and BAT First and Result WON
tossWon_batwin = Chennai[(Chennai['toss_decision'] == 'bat') & (Chennai['winner'] == 'Chennai Super Kings')]
print(tossWon_batwin)
```

```python
# Total count of CSK  Toss WON and BAT First and Result WON
tossbatwincount = len(tossWon_batwin)
tossbatwincount
```

```python
# Total count of CSK  Toss WON and BAT First and Result WON
tossbatlosscount = len(tossWon_batlost)
tossbatlosscount
```

```python
print("Count of Toss Decision: Bat and Result- Win:", tossbatwincount)
print("Count of Toss Decision: Bat and Result- LOSS:",tossbatlosscount)
print("Count of Toss Decision: Field and Result- Win :", tossfieldwincount)
print("Count of Toss Decision: Field and Result- Loss:", tossfieldlostcount)
print("Total CSK Win Count:", Chennaiwincount)
print("Total CSK Lost Count:", Chennailostcount)
```

```python
# Average Matches WON by CSK, Venue- Chennai

total_matches = len(Chennai)
total_wins = len(Chennai[Chennai['winner'] == 'Chennai Super Kings'])
average_win = total_wins / total_matches
print("Average number of matches won by Chennai Super Kings  in Chennai:", average_win)
```

```python
# Average Matches LOSS by CSK, Venue- Chennai

total_matches = len(Chennai)
total_losses = len(Chennai[Chennai['winner'] != 'Chennai Super Kings'])
average_loss = total_losses / total_matches
print("Average number of matches lost by Chennai Super Kings  in Chennai:", average_loss)
```
